```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to reduce the intensity levels of the image
def reduce_intensity_levels(image, levels):
    assert (levels & (levels - 1) == 0) and levels <= 256, "Levels must be in power of 2 and <= 256"
    factor = 256 // levels
    reduced_img = (image // factor) * factor
    return reduced_img
```

```python
# Load the image in greyscale mode
img_path = os.path.join("..", "data", "task_01", "image_01.jpg")
original_img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

if original_img is None:
    raise FileNotFoundError(f"Image not found at {img_path}")

# Intensity level variable intialization 
desired_levels = 16  #reducing to 16 intensity levels 
```

```python
# Process image down to the desired intensity levels
reduced_img = reduce_intensity_levels(original_img, desired_levels)
```

```python
output_path = os.path.join("..", "results", "task_01", f"reduced_sample_{desired_levels}_levels.png")
cv2.imwrite(output_path, reduced_img)
```

```python
# Display the results
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(original_img, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title(f"Reduced to {desired_levels} Levels")
plt.imshow(reduced_img, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
