```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Load image in color
img_path = os.path.join("..", "data", "task_02", "image_02.jpg")
img = cv2.imread(img_path, cv2.IMREAD_COLOR)

if img is None:
    raise FileNotFoundError(f"Image not found at {img_path}")
```

```python
# Function to apply averaging filter
def apply_mean_filter(image, kernel_size):
    return cv2.blur(image, (kernel_size, kernel_size))
```

```python
# Apply filters
img_3x3 = apply_mean_filter(img, 3)
img_10x10 = apply_mean_filter(img, 10)
img_20x20 = apply_mean_filter(img, 20)
```

```python
# Convert BGR to RGB for correct display with matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_3x3_rgb = cv2.cvtColor(img_3x3, cv2.COLOR_BGR2RGB)
img_10x10_rgb = cv2.cvtColor(img_10x10, cv2.COLOR_BGR2RGB)
img_20x20_rgb = cv2.cvtColor(img_20x20, cv2.COLOR_BGR2RGB)
```

```python
# Display results
titles = ['Original', '3x3 Average', '10x10 Average', '20x20 Average']
images = [img_rgb, img_3x3_rgb, img_10x10_rgb, img_20x20_rgb]

plt.figure(figsize=(20, 5))
for i in range(4):
    plt.subplot(1, 4, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
#Save outputs to results
result_dir = os.path.join("..", "results", "task_02")
os.makedirs(result_dir, exist_ok=True)

cv2.imwrite(os.path.join(result_dir, "avg_3x3.jpg"), img_3x3)
cv2.imwrite(os.path.join(result_dir, "avg_10x10.jpg"), img_10x10)
cv2.imwrite(os.path.join(result_dir, "avg_20x20.jpg"), img_20x20)
```
